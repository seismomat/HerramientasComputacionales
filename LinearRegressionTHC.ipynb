```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

```python
def get_data(N):
  x=np.linspace(0,2,N)
  y=np.zeros_like(x)
  eps=1E-1
  for i in range(N):
    y[i]=2*x[i]+3+(eps*(-1)**i)

  return x,y
```

```python
def plot_data(x,y,params):
  z=params[0]+params[1]*x

  plt.figure()
  plt.plot(x,y,'r*',label="Samples")
  plt.plot(x,z,'-k',label="Model")
  plt.title('Linear Regression')
  plt.xlabel('X Data')
  plt.ylabel('Y Data')
  plt.grid(lw=2)
  plt.legend()
  plt.show()
```

```python
def linear_regression(N=20):
  x,y=get_data(N)
  A=np.vstack([np.ones(len(x)),x]).T
  # el b es la y
  AtA=A.T@A
  Aty=A.T@y
  params=LA.solve(AtA,Aty)
  plot_data(x,y,params)
  OriData=np.array([3.0,2.0])
  Error=np.round(LA.norm(params-OriData),3)
  print('Error: ',Error*100,'%')
  return params
```

```python
linear_regression()
```
