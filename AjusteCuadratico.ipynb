```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los puntos de datos
x_points = np.array([1, 2, 3, 4, 5])  # Ejemplo de coordenadas x
y_points = np.array([2, 3, 5, 7, 11]) # Ejemplo de coordenadas y

# Crear la matriz A
A = np.vstack([ np.ones(len(x_points)),x_points,x_points**2]).T  # n x 3 matriz

# Resolver el sistema usando mínimos cuadrados
coef, residuals, rank, s = np.linalg.lstsq(A, y_points, rcond=None)
c, b, a = coef

print("Coeficiente a:", a)
print("Coeficiente b:", b)
print("Coeficiente c:", c)
```

```python
# Función parabólica ajustada
def parabola(x):
    return a * x**2 + b * x + c

# Crear una serie de puntos para graficar la parábola
x_values = np.linspace(min(x_points), max(x_points), 100)
y_values = parabola(x_values)
```

```python
# Graficar los puntos originales y la parábola ajustada
plt.scatter(x_points, y_points, color='red', label='Puntos dados')
plt.plot(x_values, y_values, label='Parábola ajustada')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolación Parabólica por Mínimos Cuadrados')
plt.legend()
plt.grid(True)
plt.show()
```
