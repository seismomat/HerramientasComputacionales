```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA

def get_data(N=20):
  x=np.linspace(0,2,N)
  eps=1E-1;y=[]
  for i in range(N):
    y.append(2*x[i]+3+((-1)**i*eps))

  y=np.array(y)

  return x,y
```

```python
def Regresion():
    x,y = get_data(30)
    # Crear la matriz A
    A = np.vstack([ np.ones(len(x)),x]).T  # n x 3 matriz
    b=y

    # Resolver el sistema usando mínimos cuadrados
    Params, residuals, rank, s = np.linalg.lstsq(A, b, rcond=None)

    print("Solución:", x)
    print("Residuos:", residuals)  # Error en mínimos cuadrados
    z=Params[0]+Params[1]*x # creo un modelo auxiliar con los valores
                        # aproximados

    plt.plot(x,y,'r*',label="Datos")
    plt.plot(x,z,'-k',lw=3,label="Ajuste lineal")
    plt.title("Regresion lineal")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(lw=1)
    plt.legend()
    plt.show()


    return Params


Regresion()
```
